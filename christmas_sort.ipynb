```python
import pandas as pd
import numpy as np
import random
```

```python
INFILE = '/Users/klogg/Desktop/maddock_family_email_2024.csv'
OUTFILE = '/Users/klogg/Desktop/maddock_family_email_2024_assignments.csv'

df = pd.read_csv(INFILE)
```

```python
df = df.loc[df['Name'] != 'Sophie'].reset_index(drop=True)
df
```

```python
while True:
    reshuffle = False
    df['Gives To'] = df['Name'].sample(frac=1).reset_index(drop=True)
    for i,x in df.iterrows():
        if x['Gives To'] == x['Name']:
            reshuffle = True
            break
        elif x['Family Group'] == df.loc[df['Name'] == x['Gives To']]['Family Group'].values[0]:
            reshuffle = True
            break
        elif x['Gave to 2023'] == x['Gives To']:
            reshuffle = True
            break
        elif x['Gave to 2022'] == x['Gives To']:
            reshuffle = True
            break
        elif x['Gave to 2021'] == x['Gives To']:
            reshuffle = True
            break
    if not reshuffle:
        break
        
df
```
